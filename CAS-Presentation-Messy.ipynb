```python
import requests
import json
from datetime import datetime
import pandas as pd
import time
```

```python
API_KEY = ''
if(False):
    ZIP_CODE = '11103'
    LOCATION = '40.7644,-73.9184'
    RADIUS = 2000  # in meters

    url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={LOCATION}&radius={RADIUS}&type=restaurant&key={API_KEY}"
    response = requests.get(url)
    restaurants = response.json().get('results', [])
    timestamp = datetime.now().strftime('%Y-%m-%d_%H-%M-%S')
    filename = f'restaurants_{ZIP_CODE}_{timestamp}.json'
    
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(restaurants, f, indent=2)
    
    print(f"Saved {len(restaurants)} restaurants to {filename}")
```

```python
if(False):
    addresses = [
        "28-30 31st St, Astoria, NY 11102",
        "25-19 30th Ave, Astoria, NY 11102",
        "43-05 Broadway, Astoria, NY 11103",
        "Sweet Afton, Astoria, NY 11103"
    ]
    
    def get_place_data(address):
        url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json"
        params = {
            "input": address,
            "inputtype": "textquery",
            "fields": "place_id",
            "key": API_KEY
        }
        response = requests.get(url, params=params)
        return response.json()
    
    def get_place_details(place_id):
        url = "https://maps.googleapis.com/maps/api/place/details/json"
        params = {
            "place_id": place_id,
            "fields": "name,formatted_address,types,website,international_phone_number",
            "key": API_KEY
        }
        response = requests.get(url, params=params)
        return response.json().get("result", {})
    
    results = []
    for addr in addresses:
        print(f"Searching: {addr}")
        data = get_place_data(addr)
        candidates = data.get("candidates", [])
        if candidates:
            place_id = candidates[0]['place_id']
            details = get_place_details(place_id)
            results.append(details)
        else:
            print("No match found.")
        time.sleep(0.2)  # Respect rate limits
    
    # === Save to timestamped JSON file ===
    timestamp = datetime.now().strftime('%Y-%m-%d_%H-%M-%S')
    filename = f'restaurants_by_address_{timestamp}.json'
    
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(results, f, indent=2)
    
    print(f"Saved {len(results)} restaurants to {filename}")
```

```python
cleaned_data = pd.DataFrame()
cleaned_data['Name'] = [restaurants[i]['name'] for i in range(len(restaurants))]
cleaned_data['Types'] = [restaurants[i]['types'] for i in range(len(restaurants))]

ALCOHOL_KEYWORDS = ['bar', 'pub', 'tavern', 'brewery', 'wine', 'cocktail', 'liquor', 'taproom']

def serves_alcohol(name, types):
    for c in ALCOHOL_KEYWORDS:
        if(c in name.lower() or c in types):
            return 1
    return 0

cleaned_data['Alcohol Flag'] = [serves_alcohol(cleaned_data['Name'][i],cleaned_data['Types'][i]) for i in range(len(cleaned_data))]
```
